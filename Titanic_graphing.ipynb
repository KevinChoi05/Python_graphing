```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df = pd.read_csv('titanic.csv')
df.head()
```

# Histogram

```python
plt.figure(figsize=(6,4))
plt.hist(df[df['Pclass'] == 1]['Fare'], alpha = 0.5, label = 'first class')
plt.hist(df[df['Pclass'] == 2]['Fare'], alpha = 0.5, label = 'second class')
plt.hist(df[df['Pclass'] == 3]['Fare'], alpha = 0.5, label = 'third class')
plt.title('Histogram of Fare by Pclass')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# Scatter plot

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df, x='Fare', y='Age', hue = df['Survived'], palette = 'muted')
plt.title('scatterplot of fare and age')
plt.show()
```

# Pie chart

```python
df0 = len(df[df['Survived']==0])
df1 = len(df[df['Survived']==1])
ratio = [df0, df1]
labels=['dead', 'Survived']
plt.figure(figsize=(6,6))
plt.pie(ratio, labels = labels, autopct='%1.1f%%')
plt.title('Pie chart among dead or survived')
plt.legend(loc ='best', bbox_to_anchor=(1,1))
plt.show()
```

# Box plot

```python
plt.figure(figsize = (6,4))
fig, ax = plt.subplots()
ax.boxplot([df[df['Pclass'] == 1]['Fare'], df[df['Pclass'] == 2]['Fare'], df[df['Pclass'] == 3]['Fare']])
plt.xticks([1,2,3], ['first class', 'second class', 'third class'])
plt.title('box plot among fare and pclass')
plt.tight_layout()
plt.show()
```
