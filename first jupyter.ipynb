```python
from sklearn.datasets import load_iris
```

```python
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
```

```python
iris['data']
```

```python
features = iris['data']
```

```python
label = iris['target']
```

```python
feature_names = iris['feature_names']
```

```python
df = pd.DataFrame(features, columns = feature_names)
```

```python
df.head(3)
```

```python
df.tail(3)
```

```python
df['target'] = label 
```

```python
df.head(4)
```

```python
df.describe()
```

```python
df0 = df[df['target'] == 0]
df1 = df[df['target'] == 1]
df2 = df[df['target'] == 2]
x = np.arange(len(df0))
```

```python
df0['sepal length (cm)']
```

# Setosa:

# Sepal length graphs:

```python
plt.figure(figsize=(6,4))
plt.hist(df0['sepal length (cm)'])
plt.title('sepal length (cm) of Iris Setosa')
plt.show()

plt.figure(figsize=(6,4))
plt.plot(df0['sepal length (cm)'])
plt.title('sepal length (cm) of Iris Setosa')
plt.show()

plt.figure(figsize=(6,4))
plt.bar(x, df0['sepal length (cm)'])
plt.title('sepal length (cm) of Iris Setosa')
plt.show

plt.figure(figsize=(6,4))
fig, ax = plt.subplots()
ax.boxplot(df0['sepal length (cm)'])
plt.title('boxplot of sepal length (cm) of Iris Setosa')
plt.show()
```
